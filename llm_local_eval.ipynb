# 2.启动本地评估
使用 gsm8k 数据集，gsm8k 是一个包含 8.5K 道的小学数学应用题的数据集，关于数据集详细说明请访问 https://huggingface.co/datasets/openai/gsm8k 地址查阅。

```python
!evalscope eval \
 --model Qwen/Qwen2.5-0.5B-Instruct \
 --datasets gsm8k arc \
 --limit 5
```


上面的评测结果表示 **Qwen2.5-0.5B-Instruct** 模型在数据集 **arc** 的子数据集 **ARC-Easy** 和 **ARC-Challenge** 的**平均准确率**分别为 **80%** 和 **40%**，评测的样本数量为 5。


在数据集 **gsm8k** 的子数据集 **main** 的**平均准确率**为 **40%**，评测的样本数量为 5。

- Model Name: 模型名称，例如 Qwen2.5-0.5B-Instruct。
- Dataset Name: 数据集名称，例如 gsm8k 或 ai2_arc。
- Metric Name: 评测指标名称，例如 AverageAccuracy (平均准确率)。
- Category Name: 类别名称，通常是 default。
- Subset Name: 数据集的子集名称，例如 main、ARC-Easy 或 ARC-Challenge。
- Num: 评测的样本数量，例如 5。
- Score: 模型在该数据集上的得分，例如 0.4 或 0.8。
